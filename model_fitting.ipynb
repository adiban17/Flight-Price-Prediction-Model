```python
# importing required libraries
import pandas as pd
import numpy as np
```

```python
# loading the cleaned dataset
df=pd.read_csv('cleaned_flight_price_data_filled.csv')
df.drop('Unnamed: 0',axis=1,inplace=True)
df.head()
```

```python
df.info()
```

```python
# dividing the dataset into value and target columns
x=df.drop('Price',axis=1)
y=df['Price']
```

```python
x.head()
```

```python
y.head()
```

```python
# picking a random seed to save our results
np.random.seed(49)

# using the ExtraTreesRegressor algorithm to train our model
from sklearn.ensemble import ExtraTreesRegressor
model=ExtraTreesRegressor()

# dividing the data  into training and testing data
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test=train_test_split(x,y)

# fitting the data into the model
model.fit(x_train,y_train)

# scoring the model based on testing data
model_score=model.score(x_test,y_test)
print("Accuracy:",model_score*100,"%")
```

```python
# saving and exporting the model
import joblib
joblib.dump(filename='flight_price_predictor.pkl', value=model)
```
